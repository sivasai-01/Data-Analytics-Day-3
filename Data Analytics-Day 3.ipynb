```python
##ggplot
from matplotlib import style
style.use('ggplot')
#plt.plot(li,li2)
days=[1,2,3,4,5]
temp=[34.77,36,80,23.6,78]

plt.plot(days,temp,color='g',linestyle='--',linewidth=20,marker='^',markersize=10)

plt.show()
```

```python
vij_temp=[63.5,45.4,61,54.5,45]
plt.plot(days,vij_temp,'m*--',linewidth=5,label="vij")
plt.plot(days,temp,color='g',linestyle='--',linewidth=5,marker='^',markersize=3,label='gnt')
plt.title('gnt-temp vs Vij_temp')
plt.legend(loc=4)
plt.show()
```

```python
vij_temp=[63.5,45.4,61,54.5,45]
plt.plot(days,vij_temp,'m*--',linewidth=5,label="vij")
plt.plot(days,temp,color='g',linestyle='--',linewidth=5,marker='o',markersize=10,label='gnt')
plt.title('gnt-temp vs Vij_temp')
plt.legend(loc=4)
plt.grid(color='r',linestyle='-',linewidth=2)
plt.show()
```

## Bar Graphs

```python
classes=['python','AI','DS',"CO",'OS']
st=[10,20,30,40,50]
st2=[60,70,80,90,100]
st3=[54,68,14,36,78]
plt.bar(classes,st)
```

```python
plt.bar(st,classes,color='black')
```

```python
plt.bar(classes,st2,width=0.5,color='white',align='edge',edgecolor='m',linewidth=2,linestyle='--')
```

```python
plt.bar(classes,st2,color='white')
plt.bar(classes,st)
plt.show()
```

```python
plt.barh(classes,st)
```

```python
plt.figure(figsize=(15,9))
plt.bar(classes,st)
```

```python
cls_index=np.arange(len(classes))
width=0.2
plt.bar(classes,cls_index)
```

```python
plt.figure(figsize=(4,6))
plt.bar(classes,cls_index,st,width,color='g')
plt.bar(classes,cls_index+width,st2,width,color='b')
plt.bar(classes,cls_index+width+width,st3,width,color='b')
plt.bar(cls_index+width,classes)
plt.yticks(np.arange(0,60,15))
```
